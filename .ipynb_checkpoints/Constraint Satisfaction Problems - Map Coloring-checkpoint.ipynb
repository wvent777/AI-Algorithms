## CSP: Map Coloring

In this programming assignment, you will be using your new understanding of **Constraint Satisfaction Problems** to color maps. As we know from the [Four Color Theorem](http://en.wikipedia.org/wiki/Four_color_theorem) any division of a plane into contiguous regions can be colored such that no two adjacent regions are the same color by using only four colors.

From the book, we know that we can translate this problem into a CSP where the map is represented as a [planar graph](http://en.wikipedia.org/wiki/Planar_graph) and the goal is to color all the nodes such that no adjacent nodes are colored the same color.

As with most AI problems, this requires us to figure out how best to represent the problem--and the solution--given the high and low level data structures and types at our disposal. For this problem, we'll settle on a Dict which contains at least two keys: "nodes" which is an *ordered* List of Strings that represents each node or vertex in the planar graph and "edges" which contains a List of Tuples that represent edges between nodes. The Tuples are of ints that represent the index of the node in the "nodes" list.

Using this system, and adding a "coordinates" key with abstract drawing coordinates of each node for NetworkX, we can represent the counties of Connecticut like so:

````python
connecticut = { "nodes": ["Fairfield", "Litchfield", "New Haven", "Hartford", "Middlesex", "Tolland", "New London", "Windham"],
                "edges": [(0,1), (0,2), (1,2), (1,3), (2,3), (2,4), (3,4), (3,5), (3,6), (4,6), (5,6), (5,7), (6,7)],
                "coordinates": [( 46, 52), ( 65,142), (104, 77), (123,142), (147, 85), (162,140), (197, 94), (217,146)]}
print(connecticut)
````

The coordinates permit us to use NetworkX to draw the graph. We'll add a helper function for this, `draw_map`, which takes our planar_map, a figure size in abstract units, and a List of color assignments in the same order as the nodes in the planar_map.  The underlying drawings are made by matplotlib using NetworkX on top of it. Incidentally, the positions just make the map "work out" on NetworkX/matplotlib.

The size parameter is actually inches wide by inches tall (8, 10) is an 8x10 sheet of paper. Why doesn't a chart cover up the whole screen then? It's adjusted by dpi. On high resolution monitors, 300 dpi with 8x10 inches might only take up a fraction of that space. Use whatever you want to make the output look good. It doesn't matter for anything else but that.

A default value for `color_assignments` is provided, `None`, that simply colors all the nodes red. Otherwise, `color_assignments` must be a `List of Tuples` where each `Tuple` is a node name and assigned color. The order of `color_assignments` must be the same as the order of `"nodes"` in the `planar_map`.

**If you are using NetworkX in your assignment, you can copy this function into module03.py**

````python
def draw_map(name, planar_map, size, color_assignments=None):
    def as_dictionary(a_list):
        dct = {}
        for i, e in enumerate(a_list):
            dct[i] = e
        return dct
    
    G = nx.Graph()
    
    labels = as_dictionary(planar_map["nodes"])
    pos = as_dictionary(planar_map["coordinates"])
    
    # create a List of Nodes as indices to match the "edges" entry.
    nodes = [n for n in range(0, len(planar_map["nodes"]))]

    if color_assignments:
        colors = [c for n, c in color_assignments]
    else:
        colors = ['red' for c in range(0,len(planar_map[ "nodes"]))]

    G.add_nodes_from(nodes)
    G.add_edges_from(planar_map[ "edges"])

    plt.figure(figsize=size, dpi=100)
    plt.title(name)
    nx.draw(G, node_color = colors, with_labels = True, labels = labels, pos = pos)
    
#    plt.savefig(name + ".png")
````

<a id="check_finished"></a>
## check_finished

This function evaluates if the current solution is complete by checking if the solution has assigned values to every variable in the problem. 

* **solution** Tuple[int, int]: the starting location of the bot, `(x, y)`.
* **variables** List[List[str]]: the actual context for the navigation problem.

**returns** Boolean

````python
def check_finished(solutions:dict, variables:list[str])->bool:
    if len(solutions) == len(variables):
        return True
    else:
        return False
````

````python
test_solution = {'Fairfield': 'red', 'Litchfield': 'blue'}
test_variables = ['Fairfield', 'Litchfield']

assert check_finished(test_solution, test_variables) ==  True

test_solution = {'Fairfield': 'red', 'Litchfield': 'blue'}
test_variables = ['Fairfield', 'Litchfield', "Fairfax"]

assert check_finished(test_solution, test_variables) ==  False

test_solution = {'Fairfield': 'red', 'Litchfield': 'blue', 'fairfax': 'yellow'}
test_variables = ['Fairfield', 'Litchfield', 'fairfax']

assert check_finished(test_solution, test_variables) ==  True
````

<a id="select_static_variable"></a>
## select_static_variable

Variable Selction Heuristic that returns the first unassigned variable in list

* **csp** List[List[str]]: the actual context for the navigation problem.
* **solution** Tuple[int, int]: the starting location of the bot, `(x, y)`.

**returns** str -  Variable

````python
def select_static_variable(solution, csp, trace=True):
    unassigned_var = [x for x in csp[0] if x not in solution]
    variable = unassigned_var[0]
    if trace:
        print("  Selected Variable: {0}".format(variable))
    return variable
````

<a id="select_static_domain"></a>
## select_static_domain

Domain-value selection heurisitc that returns the value in a static ordering, e.g. (1, 2, 3, 4)

* **csp** List[List[str]]: 
* **variable** str: the variable
* **assignment** str: the assignment for the variable

**returns** list

````python
def select_static_domain(variable, assignment, csp, trace=True):
    for domain_values in csp[1][variable]:
        if trace: print ("  Selected value {0} for {1}".format(domain_values, variable))
        yield domain_values
````

<a id="check_consistency"></a>
## check_consistency

Checks if there is any conflicts between the value being assigned to the variable and the current assignments

* **value** str
* **variable** str: the variable
* **assignment** str: the assignment for the variable
* **constraints**: the neighbors that the variable is constrained to 

**returns** Boolean

````python
def check_consistency(variable, value, assignment, constraints, trace=True):
    if assignment is None: return False
    for v in constraints:
        if v in assignment:
            if value is assignment[v]:
                if trace: 
                    print ("  {0} : {1} conflicts with {2} :{3}".format(variable,
                                                                       value, v, assignment[v]))
                return False
    return True
````

<a id="backtrack"></a>
## backtrack

recursive backtrack function, that works by using depth first search to assign values to variables based on certain heuristics. If value can't be assigned then it backtracks to previus successful assignment. Continues to do this until a solution is found if there is one. 

* **csp** tuple(Variables, Domain, Constraints): 
* **assignment** dict: empty dictionary that holds the current solution
* **parameters** tuple: heuistics 

**returns** Boolean, str 

````python
def backtrack(assignment:str, csp:tuple, parameters:tuple, trace=True)->str:
    select_unassigned_variable, order_domain_values, inference = parameters
    if trace: 
        print ("Solution: {0}".format(assignment))
    if check_finished(assignment, csp[0]): 
        return True, assignment
    variable = select_unassigned_variable(assignment, csp, trace=trace)
    for value in order_domain_values(variable, assignment, csp, trace):
        if check_consistency(variable, value, assignment, csp[2][variable], trace=trace):
            assignment[variable], m_csp = value, deepcopy(csp)
            if inference(csp, variable, value, assignment, trace=trace):
                result, assignment = backtrack(deepcopy(assignment), csp, parameters, trace=trace)
                if result: 
                    return True, assignment
            del assignment[variable]
            csp = m_csp
    return False, assignment
````

<a id="backtrack_search"></a>
## backtrack_search

* **csp** tuple(Variables, Domain, Constraints): 
* **parameters** tuple

**returns** Boolean


````python
def backtrack_search(csp, parameters, trace=True):
    solution = dict()
    return backtrack(solution, csp, parameters, trace=trace)
````

<a id="get_adjacents"></a>
## get_adjacents

* **nodes** 
* **edges**

**returns** Boolean


````python
def get_adjacents(nodes, edges):
    adjacency_list = dict()
    for node in nodes:
        aList = set()
        [aList.add(nodes[x]) for x, y in edges if nodes[x] != node and nodes[y] == node]
        [aList.add(nodes[y]) for x, y in edges if nodes[y] != node and nodes[x] == node]
        adjacency_list[node] = aList
    return adjacency_list
````

<a id="mapping"></a>
## mapping


* **data** dictionary: The dictionary that is given for the planar mapping

**returns** tuple

````python
def mapping(data):
    nodes = data['nodes']
    edges = data['edges']
    return (nodes, edges)
````

<a id="forward_checking"></a>
## forward_checking

Forward Checking function deletes all values from the adjacent nodes' domains

* **csp** tuple(Variables, Domain, Constraints)
* **assignment** str
* **parameters** tuple

**returns** Boolean

````python
def forward_checking(csp, variable, value, assignment, trace=True):
    unassigned = [y for y in csp[2][variable] if y not in assignment]
    for x in unassigned:
        if value in csp[1][x]:
            if trace:
                print("  Removing: {0} from {1}".format(value, x))
            csp[1][x] =  [c for c in csp[1][x] if c is not value]
        if len(csp[1][x]) == 0:
            return False
    return True
````

<a id="mrv"></a>
## mrv - Minimum Remaining Values


**returns** str

````python
def mrv(solution, csp, trace=True):
    unassigned_var = [ x for x in csp[0] if x not in solution]
    var_domain_len = [(x, len(csp[1][x])) for x in unassigned_var]
    variables = sorted(var_domain_len, key=itemgetter(1))
    if trace:
        print("  Selected Variable: {0}, Value Count: {1}".format(variables[0][0]
                                                                  , variables[0][1]))
    return variables[0][0]
````

<a id="lcv"></a>
## lcv - Least Constraining Values


**returns** str


````python
def lcv(variable:str, assignment:str, csp:tuple, trace=True)->str:
    neighbors = csp[2][variable]
    counts = []
    for value in csp[1][variable]:
        count = 0 
        for neighbor in neighbors:
            if value in csp[1][neighbor]:
                count = count + 1 
            counts.append((value,count))
    sorted_count = sorted(counts, key=itemgetter(1), reverse=True)
    for value, count in sorted_count:
        if trace:
            print("  Selected Value: {0} for {1}".format(value, variable))
        yield value
````

````python
def color_map(planar_map:dict, color_list:list[str], parameters:tuple(str), trace=False)->list:
    variables = planar_map[0]
    domains = dict([(x,color_list) for x in variables])
    constraints =  get_adjacents(planar_map[0], planar_map[1])
    result, solution = backtrack_search((variables, domains, constraints), parameters, trace=trace)
    if not result:
        print("no solution")
    else:
        return list(solution.items())
#     return [(n, "red") for n in planar_map["nodes"]]
````
